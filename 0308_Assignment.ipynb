[03] DataFrame과 리스트,딕셔너리,넘파이 ndarray 상호 변환
 - DataFrame은 나머지와 다르게 칼럼명을 가지고 있음
 - DataFrame 들은 기본적으로 2차원이기에, 2차원 이하의 데이터들만 변환 가능
  - 1차원 기반 형성시, 칼럼이 하나 필요
  - 2차원 기반 형성시, 형태에 따라 필요한 칼럼명 수가 달라짐

```python
#1차원 형태의 리스트와 넘파이 ndarray 부터 변환
import numpy as np

col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

#리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1,columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

#넘파이 ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1,columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
```

```python
#2차원 list, ndarray 형태를 기반으로 DataFrame 생성

#3개의 칼럼명이 필요함
col_name2 = ['col1','col2','col3']

#2행 x 3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환
list2 = [[1,2,3],
         [11,12,13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape)
df_list2 = pd.DataFrame(list2,columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n:', df_list2)
df_array2 = pd.DataFrame(array2,columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
print(array2)
```

[09] apply lambda 식으로 데이터 가공
 - lambda 식은 파이썬에서 함수형 프로그래밍을 지원하기 위해서 만들었다
 - ' : ' 의 왼쪽에 있는 x는 입력인자, 오른쪽은 입력 인자의 계산식(=반환 값)이다
 - 여러 개의 값을 입력 인자로 사용해야 할 경우는 map() 함수를 결합해서 사용한다

```python
#몇 가지 예시
def get_square(a):
  return a**2

print('3의 제곱은:', get_square(3))
```

```python
#위의 예시를 lambda 식으로 변형 했을 때
```

```python
lambda_square = lambda x : x ** 2
print('3의 제곱은:',lambda_square(3))
```

```python
a=[1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
#apply에 lambda식을 적용해 가공하기
titanic_df['Name_len']= titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name','Name_len']].head(3)
```

- lambda 식에서 if 절 사용시 주의점
 - if 식보다 반환 값을 앞에 기술해야 함</br>
  ex) lambda x : if x <=15 'Child' else 'Adult' (X)</br>
      lambad x : 'Child' if x<=15 else 'Adult' (O)

```python
#나이가 15세 미만이면 'Child', 그렇지 않으면 'Adult' 구분
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <=15 else 'Adult' )
titanic_df[['Age','Child_Adult']].head(8)
```

- if 와 else 만 지원하고 else if 는 지원하지 않는다.</br> else if 를 이용하려면 else절에 또 다른 else절을 내포해야 한다
- 이 역시도 반환 값이 식보다 먼저 나왔음

```python
#15세 이하면 Child, 16~60은 Adult 아니면 Elderly로 분류
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x <= 60 else
                                                                                  'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성.
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'

    return cat

# lambda 식에 위에서 생성한 get_category( ) 함수를 반환값으로 지정.
# get_category(X)는 입력값으로 ‘Age’ 컬럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
```
